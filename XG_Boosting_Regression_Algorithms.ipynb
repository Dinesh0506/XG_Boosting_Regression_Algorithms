```python
#import the libary and pandas is used to numerical data fuction 
import pandas  as pd
```

```python
#usind pandas to read a data and assign to dataset varible  &print the data having dataset variablle
dataset=pd.read_csv('insurance_pre.csv')
dataset
```

```python
#convert the categorical data to numerical data 
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#split the input data and print the data 
indepentent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
indepentent
```

```python
#split the output data and print the data
depentent=dataset[["charges"]]
depentent
```

```python
#split the train data and test data using function train_test_split undert sklearn function 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indepentent,depentent, test_size = 1/3, random_state = 0)
```

```python
#inthis model creation of xgboost
import xgboost as xg 
regressor=xg.XGBRegressor(objective ='reg:linear',n_estimators = 10, seed = 123)
regressor=regressor.fit(X_train,y_train)
```

```python
 y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
